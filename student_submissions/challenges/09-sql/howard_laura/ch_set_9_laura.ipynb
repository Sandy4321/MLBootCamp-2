5) What category appears in the most orders?

7) Which employee made the most sales (by total price
> Answer = Margaret Peacock

```sql
SELECT FirstName, LastName, SUM(TotalCost) AS AllSales FROM Orders
JOIN Employees ON Orders.EmployeeID = Employees.EmployeeID
JOIN
  (
  SELECT OrderID, SUM(Price) * Quantity AS TotalCost
  FROM OrderDetails
  JOIN Products ON Products.ProductID = OrderDetails.ProductID
  GROUP BY OrderDetails.OrderID
  ORDER BY TotalCost DESC
  ) CostPerOrder
ON CostPerOrder.OrderID = Orders.OrderID
GROUP BY Employees.EmployeeID
ORDER BY AllSales DESC LIMIT 1;
```

9) Which supplier of three or more products has the highest average product price?
> Answer = Tokyo Traders

```sql
SELECT SupplierName, AVG(Price) AS AverageProductPrice, COUNT(*) AS ProductCount
FROM Products
JOIN Suppliers ON Products.SupplierID = Suppliers.SupplierID
GROUP BY Products.SupplierID
HAVING ProductCount >= 3
ORDER BY AverageProductPrice DESC LIMIT 1;
```
