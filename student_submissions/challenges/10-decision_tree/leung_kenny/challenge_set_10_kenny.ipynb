```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
import graphviz

%matplotlib inline
```

# Challenge 1

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data'
```

```python
df = pd.read_csv(url,header=None)
df = df.replace(['y','n'],[1,0])
```

```python
issues = dict()
for issue in range(1,17):
    issues[issue] = (df.groupby(issue).count()[0][0], 
                     df.groupby(issue).count()[0][1], 
                     df.groupby(issue).count()[0][0]/(df.groupby(issue).count()[0][0]+df.groupby(issue).count()[0][1]))
    
    df[issue] = df[issue].replace('?',issues[issue][2])
```

```python
X = df.loc[:,1:16]
y = df.loc[:,0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=4444)
```

```python
tree1 = tree.DecisionTreeClassifier()
tree1.fit(X_train,y_train)
```

```python
dot_data = tree.export_graphviz(tree1, out_file=None,feature_names=X.columns,class_names=tree1.classes_,filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph
```
