```python
%%time
ua = UserAgent()

flight_ls_20 = []

for i in range(len(airport_list)):
    for j in range(len(airport_list)):
        if i != j:
            a = airport_list[i]
            b = airport_list[j]
            url = "http://flightaware.com/live/findflight?origin={}&destination={}"

            path = url.format(a,b)
#             path = 'http://flightaware.com/live/findflight?origin=ZSHC&destination=KSFO'
    
            user_agent = {'User-agent': ua.random}
            response = requests.get(path, headers = user_agent)


            if response.status_code == 200:
                print(a, b)

                page = response.text

                soup = BeautifulSoup(page,'lxml')
                results = soup.find('table', id="Results")
                has_co = len(results.find_all('th'))
                row_total = len(results.find_all('tr'))

                if has_co ==8: 
                    find_flight_number(path, row_total, a, b)
                else:    
                    for row in results.find_all('tr', url = True):
                        x = row['url']
                        x = x[x.find('id')+3:x.find('-')]
                        flight_ls_20.append(x)
                    
            sleep(2+2*random.random())
```
